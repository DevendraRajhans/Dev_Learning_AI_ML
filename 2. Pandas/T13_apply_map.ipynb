```python
import pandas as pd
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha"],
    "Marks": [85, 90, 35, 72],
    "City": ["Pune", "Mumbai", "Delhi", "Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. map() 
# > map works on series only (not on Dataframes)

# Converting names of city (if names doesn't exist it puts NaN there)
df["City"] = df["City"].map({
    "Pune": "MH",
    "Mumbai": "MH",
    "Delhi": "DL"
})
print(df)
```

```python
df["Grade"] = df["Marks"].map(lambda x: "Pass" if x >= 40 else "Fail")
df
```

```python
# 2. apply()
# works on Series and DataFrames both

# a) on series
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
df
```

```python
# b) on rows (axis=0)
# when logic needs multiple columns. 
df["Scholarship"] = df.apply(
    lambda row: "Yes" if row["Marks"] > 80 and row["City"] == "MH" else "No",
    axis=1
)
print(df)
```

```python
# c) on columns (axis=1)
df.apply(max) # max of all cols
```

```python
# 3. applymap()
# works on entire DataFrame
# This example is for basic idea
df.applymap(lambda x: x * 2)
df
```
