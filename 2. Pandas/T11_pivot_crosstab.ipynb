```python
import pandas as pd
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha", "Kiran", "Rohit"],
    "Age": [23, 53, 32, 36, 42, 26],
    "Department": ["IT", "IT", "HR", "HR", "IT", "HR"],
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Pune", "Delhi"],
    "Salary": [50000, 60000, 45000, 48000, 52000, 47000]
}

df = pd.DataFrame(data)
print(df)
```

--------------------------------------------- 

✅ A) What is Pivot Table?
> It summarizes data using:
- rows (index)
- columns
- values
- aggregation functions

```python
# 1. Avg salary by department
pd.pivot_table(
    df,
    index="Department",
    values="Salary",
    aggfunc="mean"
)
```

```python
# 2. Avg salary by dept & city
pd.pivot_table(
    df,
    index="Department",
    columns="City",
    values="Salary",
    aggfunc="mean"
)
```

```python
# 3. Multiple aggregation
pd.pivot_table(
    df,
    index="Department",
    values="Salary",
    aggfunc=["mean", "min", "max"]
)
```

```python
# 4. Fill missing values in pivot table 
pd.pivot_table(
    df,
    index="Department",
    columns="City",
    values="Salary",
    aggfunc="mean",
    fill_value=0
)
```

```python
# 5. pivot table using multiple values
pd.pivot_table(
    df,
    index="Department",
    values=["Salary", "Age"],
    aggfunc=["mean", "sum"]
)
```

✅ B) What is Crosstab?
>It is used to count frequency of combinations.

```python
# 1) Count employees by Department
pd.crosstab(df["Department"], df["City"])
```

```python
# 2) Count with margins
pd.crosstab(df["Department"], df["City"], margins=True)
```

```python
# 3) Normalize (percentage)
pd.crosstab(df["Department"], df["City"], normalize=True)
```
