```python
import pandas as pd
import numpy as np
```

```python
# Missing values 
# Nan or None
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", None, "Sneha"],
    "Age": [20, np.nan, 22, 21, np.nan],
    "City": ["Pune", "Mumbai", None, "Delhi", "Pune"],
    "Marks": [85, 90, np.nan, 92, 88]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. finding null values
df.isnull() # true if null
```

```python
df.isnull().sum()
```

```python
df.notnull() # true if not null
```

```python
df.notnull().sum()
```

```python
# 2. droping null values
df1 = df.dropna() # drops rows with null values
print(df1) 
```

```python
df2 = df.dropna(axis=1) # drops column with null values
print(df2)
# Empty as all columns have null values
```

```python
df3 = df.dropna(how="all")
print(df3)
#drops only if all are null
```

```python
df4 = df.dropna(subset=["Age"])
print(df4)
# drops only if age is null
```

```python
# 3. Filling null values
df1 = df.fillna(0)
print(df1)
```

```python
df1 = df
df1['Age'] = df['Age'].fillna(21)
print(df1)
```

```python
df1['Marks'] = df1['Marks'].fillna(df1['Marks'].mean())
print(df1)
```

```python
df1['City'] = df1['City'].fillna(df1['City'].mode()[0]) # df.mode() gives series. So, we need to take 0th index.
print(df1)
```

```python
# 4. Replace
# If some values are not exact null values, then we need to change it to NaN format
# values can be (Na, -1, missing, etc.)
# for this we use replace
df.replace(-1, np.nan, inplace=True) 
df2 = df.replace('na', np.nan, inplace=False)
# inplace=True → changes df directly (no reassignment needed)
# inplace=False → returns a new modified copy
```

```python
# 5. forward fill & backward fill
data = {
    "Name": ["Dev", "Rahul", "Amit", None, "Sneha"],
    "Age": [20, np.nan, 22, 21, np.nan],
    "City": ["Pune", "Mumbai", None, "Delhi", "Pune"],
    "Marks": [85, 90, np.nan, 92, 88]
}

df = pd.DataFrame(data)
print(df)
print("\nForward fill: \n", df.ffill())  # fills NaN with previous value
print("\nBackward fill: \n", df.bfill()) # fills NaN with next value
```

```python
# 6. interpolate
# filling NaN with estimating them with nearby values
data = {
    "col1":[1, 2, 3],
    "col2":[10, None, 30]
}
df = pd.DataFrame(data)
print(df)
df['col2'] = df['col2'].interpolate()
print("\nAfter Interpolation:")
print(df)
# It estimated value 20 by analysing a seq 
```
