```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha", "Kiran"],
    "Age": [20, 21, 22, 19, 21],
    "City": ["Pune", "Mumbai", "Nashik", "Delhi", "Pune"],
    "Marks": [85, 90, 88, 92, 70]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. sort_values (sort by column values)
df.sort_values("Marks") #default ascending
```

```python
df.sort_values("Name", ascending=False) #descending (Z -> A)
```

```python
# 2. multiple columns 
df.sort_values(["City", "Marks"], ascending=[True, False]) 
# first sort 'City' then 'Marks' 
# ('City' -> ascending)  then ('Marks' -> descending)
```

```python
# 3. sort_index
df.sort_index() # ascending
```

```python
df.sort_index(ascending=False) # descending
```

```python
df.sort_index(inplace=True) # sorting & saving
```

```python
# 4. sorting with missing values
df.sort_values("Name", na_position="first") # if NaN exist, then arranged  at first
```
