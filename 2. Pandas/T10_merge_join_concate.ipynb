```python
import numpy as np
import pandas as pd
```

```python
# Employees table
emp = pd.DataFrame({
    "EmpID": [1, 2, 3, 4],
    "Name": ["Dev", "Rahul", "Amit", "Sneha"],
    "DeptID": [101, 102, 101, 103]
})

# Departments table
dept = pd.DataFrame({
    "DeptID": [101, 102, 104],
    "DeptName": ["IT", "HR", "Finance"]
})

print(emp)
print(dept)
```

```python
# A) Merge
# pd.merge(left, right, on="column", how="type")
print("\nINNER:")
print(pd.merge(emp, dept, on="DeptID", how="inner"))
print("\nOUTER:")
print(pd.merge(emp, dept, on="DeptID", how="outer"))
print("\nLEFT:")
print(pd.merge(emp, dept, on="DeptID", how="left"))
print("\nRIGHT:")
print(pd.merge(emp, dept, on="DeptID", how="right"))
```

```python
# merge with different column name
pd.merge(
    emp,
    dept,
    left_on="DeptID",
    right_on="DeptID", # here you can use different one
    how="left"
)
```

```python
# B) Join
# Easier when index is the key 
# (Here we created key(DeptID) as index for both first)
emp_idx = emp.set_index("DeptID")
dept_idx = dept.set_index("DeptID")
emp_idx.join(dept_idx)
```

```python
# C) Concatenation
df1 = pd.DataFrame({
    "Name": ["Dev", "Rahul"],
    "Marks": [85, 90]
})

df2 = pd.DataFrame({
    "Name": ["Amit", "Sneha"],
    "Marks": [88, 92]
})

print(df1)
print()
print(df2)
```

```python
# 1. row-wise concate
pd.concat([df1, df2])
```

```python
# 2. column-wise concate
pd.concat([df1, df2], axis=1)
```

```python
# 3. Reset index
pd.concat([df1, df2], ignore_index=True)
```
