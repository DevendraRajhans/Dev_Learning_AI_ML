```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha"],
    "Age": [23, 12, 55, 32],
    "Marks": [85, 90, 35, 72],
    "City": ["Pune", "Mumbai", "Delhi", "Pune"]
}

df2 = pd.DataFrame(data)
print(df2)
```

```python
# 3. avoid apply() when possible
# slow
df2["Result"] = df2["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
print(df2)
print()
# Fast (vectorized)
df2["Result"] = df2["Marks"] >= 40
print(df2)
```

```python
# 4. use vectorized operations
df2["Total"] = df2["Marks"] * 1.1
df2
```

```python
# 5. filter early (reduce data size)
df2 = df2[df2['Marks'] > 50]
df2
```

```python
# 6. Use query() for Readibility
df2.query("Marks > 50 and Age >= 20")
```

```python
# 7. Use sort and groupby efficiently
df2.sort_values("Marks")
df2.groupby("City")["Marks"].mean()
```
