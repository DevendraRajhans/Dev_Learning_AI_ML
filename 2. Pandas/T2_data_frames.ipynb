```python
# Creating dataframes from different methods
import pandas as pd
```

```python
# 1. dictionary
data = {
    "name" :['as', 'sd', 'gd', 'yr'],
    "age" :[1, 2, 3, 4]
}
df1 = pd.DataFrame(data)
print(df1)
```

```python
# 2. List of Lists 
data = [
    ["sar", "par", "mar"],
    [2, 5, 7],
    ["pune", "mumbai", "delhi"]
]
df = pd.DataFrame(data, columns=['name', 'age', 'city'])
print(df)
```

```python
# 3. list of dictionaries
data = [
    {"Name": "Dev", "Age": 20, "City": "Pune"},
    {"Name": "Rahul", "Age": 21, "City": "Mumbai"},
    {"Name": "Amit", "Age": 22, "City": "Nashik"}
]

df2 = pd.DataFrame(data)
print(df2)
```

```python
# 4. NumPy arrays 
import numpy as np
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
df = pd.DataFrame(arr)
print(df)
```
