```python
import pandas as pd
```

```python
# 1. Automatically recognizes multiple formats
pd.to_datetime(["2026-02-3", "03/feb/2026", "feb 3, 2026", "3 feb 2026", "2026-02-3 14:30"], format="mixed")
```

```python
# 2. Specify date format
pd.to_datetime("15-01-2024", format="%d-%m-%Y")
```

```python
# 3. Avoid invalid or missing values
pd.to_datetime(["2026-02-03", "Bad_date"], errors="coerce")
# It becomes Nat (Not a Time)
```

```python
# 4. convert epoch time to date
pd.to_datetime(1700000000, unit="s")
```

```python
# 5. Treating timezone as UTC (Coordinated Universal Time)
pd.to_datetime("2024-01-15 12:00", utc=True)
```

```python
# 6. combine date from diff tables
# pd.to_datetime(df[["year", "month", "day"]])
```

```python
# 7. Return type control
print(type(pd.to_datetime("2024-01-15")))          # pandas.Timestamp
print(type(pd.to_datetime(["2024-01-15"])))        # pandas.DatetimeIndex
```

8. With infer_datetime_format=True
> pd.to_datetime(df["date"], infer_datetime_format=True)


Pandas:

- Looks at the first few values

- Detects a format (example: %Y-%m-%d)

- Applies that same format to all rows

- This makes parsing much faster.
