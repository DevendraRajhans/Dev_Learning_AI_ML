5. Outlier Detection (IQR Method)
> Q1 = df["Salary"].quantile(0.25)

> Q3 = df["Salary"].quantile(0.75)

> IQR = Q3 - Q1

> lower = Q1 - 1.5 * IQR

> upper = Q3 + 1.5 * IQR

> outliers = df[(df["Salary"] < lower) | (df["Salary"] > upper)]


Decision:
- Remove
- Cap
- Keep (business logic)
