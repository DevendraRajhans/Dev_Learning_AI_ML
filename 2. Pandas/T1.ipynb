```python
print(s["b"])
print(s.iloc[-1]) # by position
print(s.loc["c"]) # by label
print(s[["a", "c"]]) #multiple elements
print(s[::-1]) # slicing
print(s * 10)
```

```python
# DataFrame

# DataFrame
data = {
    "Name":['dev', 'saru', 'amu', 'pru', 'adi'],
    "age" :[20, 40, 33, 19, 24],
    "Num" :[17, 18, 15, 13, 21]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df["Name"])
print(df[["age", "Num"]])
```

```python
print(df.iloc[3]) # by position

print(df.loc[3]) # by label
```
