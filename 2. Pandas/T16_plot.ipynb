```python
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 350, 300, 450, 500],
    "Profit": [50, 80, 70, 120, 150]
}

df = pd.DataFrame(data)
print(df)
```

```python
# line plot
df.plot(x="Month", y=["Sales", "Profit"])
```

```python
# 2. bar plot
df.plot(kind="bar", x="Month", y="Sales")
```

```python
df.plot(kind="barh", x="Month", y="Sales") #horizontal
```

```python
# 3. histogram
df["Sales"].plot(kind="hist", bins=5)
```

```python
# 4. box plot
df[["Sales", "Profit"]].plot(kind="box")
```

```python
# 5. scatter plot
df.plot(kind="scatter", x="Sales", y="Profit")
```

```python
# 6. Area plot
df.plot(kind="area", x="Month", y=["Sales", "Profit"])
```

```python
# Pie chart
df["Sales"].plot(kind="pie", autopct="%1.1f%%")
```
