```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Student Name ": [" Dev ", "Rahul", "Amit", "Dev", "Sneha"],
    "AGE": ["20", "21", "22", "20", "19"],
    "City": ["pune", "Mumbai ", "NASHIK", "pune", " Pune "],
    "Marks": [85, 90, 88, 85, 92]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. Renaming
df = df.rename(columns={
    "AGE":"Age",
    "Student Name ":"StudentName"
})
print(df)
```

```python
# 2. Changing datatypes (astype)
print(df.dtypes)
```

```python
df['Age'] = df['Age'].astype(int)
print(df.dtypes)
```

```python
# 3. Duplicates
print(df.duplicated()) #check duplicates
df.drop_duplicates() #remove duplicates
df.drop_duplicates(subset="Age") #remove duplicates from age
```
