```python
import pandas as pd
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha"],
    "Age": [20, 21, 22, 19],
    "Marks": [85, 90, 88, 92]
}

df = pd.DataFrame(data)
print(df)
```

```python
# A) Adding column
# 1. adding constant column
df['Country'] = "India"
print(df)
```

```python
# 2. add with calculation
df['Marks_Bonus'] = df['Marks']+5
print(df)
```

```python
# 3. using condition { .apply() }
df['Result'] = df['Marks'].apply(lambda x: "Pass" if (x>=40) else "Fail")
df
```

```python
# 4. using multiple columns 
df["Message"] = df['Name'] + " scored " + df["Marks"].astype(str) + " Marks."
df
```

```python
# B) Updating column
df['Marks'] = df['Marks']-5
df
```

```python
# C) Removing columns { .drop() }
# 1. remove one col
df.drop("Country", axis=1)
```

```python
# 2. remove multiple cols
df.drop(["Country", "Marks_Bonus"], axis=1)
```

```python
# 3. remove Permanently 
df.drop("Country", axis=1, inplace=True)
df
```

```python
# D) Inserting column
# df.insert(position, "col_name", values)
values = ["Pune", "Mumbai", "Nashik", "Pune"]
df.insert(2, "City", values)
df
```

```python
# E) Renaming 
df.rename(columns={ "Marks":"Score" }, inplace=True)
df
```
