```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha", "Kiran", "Rohit"],
    "Age": [21, 23, 43, 28, 25, 32],
    "Department": ["IT", "IT", "HR", "HR", "IT", "HR"],
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Pune", "Delhi"],
    "Salary": [50000, 60000, 45000, 48000, 52000, 47000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. Basic single aggregation
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Department")["Salary"].sum()
```

```python
df.groupby("Department")["Name"].count()
```

```python
# 2. Multiple Aggregation ( .agg() )
df.groupby("Department")["Salary"].agg(["min", "max", "std", "mean", "sum"])
```

```python
# Renaming aggregation cols
df.groupby("Department").agg(
    min_salary = ("Salary", "min"),
    max_salary = ("Salary", "max"),
    mean_salary = ("Salary", "max"),
    sum_of_salary = ("Salary", "max")
)
```

```python
# 3. Grp on multiple column
df.groupby(["Department", "City"])["Salary"].mean() # first Dept --> then City
```

```python
# 4. Multiple columns selected
df.groupby("Department")[["Salary","Age"]].agg(["mean", "max"])
```

```python
# 5. with filtering 
df.groupby("Department")["Age"].mean().loc[lambda x: x>25]
```

```python
# 6. with sorting
df.groupby("Department")["Age"].mean().sort_values(ascending=True)
```
