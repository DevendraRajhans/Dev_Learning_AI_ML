```python
import pandas as pd
```

```python
data = {
    "Name": ["Dev", "Rahul", "Amit", "Sneha", "Kiran"],
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Pune"],
    "Department": ["IT", "HR", "IT", "HR", "IT"],
    "Grade": ["A", "B", "A", "C", "B"]
}

df = pd.DataFrame(data)
print(df)

df.memory_usage(deep=True)
```

```python
# 1. Changing to category
df["City"] = df["City"].astype("category")
df["Department"] = df["Department"].astype("category")

df.dtypes
```

- Why Use Categorical Data?

✅ Uses less memory
✅ Faster operations
✅ Better for repeated values

```python
df.memory_usage(deep=True) #less memory than above
```

```python
# value_counts
df['City'].value_counts()
```

```python
# 2. category properties : 
df["City"].cat.categories
# -- Key points
# Categories are stored once, not per row
# Order matters (especially if ordered=True)
```

```python
df["City"].cat.codes
# -- An integer array where each value is the index position of the category for that row.
# - Using the categories above:

# Category | Code
# ----------------
# Delhi	   |   0
# Mumbai   |   1
# Pune	   |   2
```

```python
# 3. Rename category
df["Grade"] = df["Grade"].astype("category")
df["Grade"] = df["Grade"].cat.rename_categories({
    "A": "Excellent",
    "B": "Good",
    "C": "Average"
})
df
```

```python
# 4. Reorder category
df["Grade"] = df["Grade"].cat.reorder_categories(
    ["Excellent", "Good", "Average"],
    ordered=True
)
df.sort_values('Grade')
```

```python
# 5. Adding & removing category
df["Grade"] = df["Grade"].cat.add_categories(["Poor"])
df['Grade'].cat.categories
```

```python
df['Grade'] = df['Grade'].cat.remove_unused_categories()
df['Grade'].cat.categories
```
