```python
df = pd.read_csv('VideoGames_sales.csv')
df.head()
```

```python
# 1. Bar charts
df['Genre'].unique()
```

```python
counts = df['Genre'].value_counts()
```

```python
counts
```

```python
x_bar = counts.index
y_bar = counts.values
```

```python
# 2. Bar chart
plt.bar(x_bar, y_bar)
plt.show()
```

```python
plt.bar(x_bar, y_bar)
plt.xticks(rotation=90) # x ticks are rotated 
plt.show()
```

```python
# 3. Count plot
sns.countplot(x='Genre', data=df, order=df['Genre'].value_counts().index, color='red')
plt.show()
```

```python
# 4. pie chart

sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales
```

```python
region_sales = sales.T.sum(axis=1)
region_sales
```

```python
plt.pie(region_sales, labels=region_sales.index, startangle=90, explode=[0,0,0,0.2], autopct='%1.1f%%')
plt.show()
```

```python
# 5. histogram
plt.hist(df['Year_of_Release'])
plt.show()
```

```python
# 6. Hist plot
sns.histplot(df['Year_of_Release'], bins= 10, binwidth=3)
plt.show()
```

```python
# 7. Kernel Density Plot
sns.kdeplot(df['Year_of_Release'])
plt.show()
```

```python
print(df['Year_of_Release'].min())
print(df['Year_of_Release'].max())
```

```python
# 8. Box plot
plt.boxplot(df['Year_of_Release'].dropna())  # drop null values
plt.show()
```
