```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("VideoGames_sales.csv")
```

```python
sns.set_theme(style = "white")
sns.lineplot(data = df, x = "Year_of_Release", y = "NA_Sales", errorbar=None) 
plt.show()
```

```python
sns.set_theme(style = "dark")
sns.lineplot(data = df, x = "Year_of_Release", y = "NA_Sales", errorbar=None) 
plt.show()
```
