```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(1, 20)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)
```

```python
plt.plot(x, y1)
plt.show()
```

```python
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y3)
plt.show()
```

```python
x = np.linspace(-10, 10, 20) # 20 values between -10 to 10
y4 = x**2
plt.plot(x, y4)
plt.show()
```

```python
# you can plot graph with symbols
plt.plot(x, y4, "g+") # green colour and + symbol
```

```python
plt.plot(x, y4, color='red', linestyle='--', marker='x')
plt.grid(True)
```
