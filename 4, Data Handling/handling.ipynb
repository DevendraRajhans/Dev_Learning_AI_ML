```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("nba.csv")
df.isnull().sum()
```

```python
# missing values 
# first check for outliers 
sns.histplot(df['Salary'], bins= 20, kde=True)
plt.show()
```

```python
# if outliers present 
df['Salary'].fillna(df['Salary'].median())
# if not present
df['Salary'].fillna(df['Salary'].mean())
# if categorical data 
df['Salary'].fillna(df['Salary'].mode()[0])
# if very big data and less missing values
df['Salary'].dropna() # this removes rows directly
print('done')
```
