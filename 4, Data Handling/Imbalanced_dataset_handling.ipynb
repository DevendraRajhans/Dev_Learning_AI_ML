```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Amount": np.random.randint(10, 5000, 1000),
    "Class": ['A']*990 + ['B']*10
})

df
```

```python
# Here this is called as imbalanced dataset 
# where 99% data is saying there is A class , therefore ML model cannot predict correctly
df['Class'].value_counts()
```

```python
# so we need to take only 10 class A rows for ML model
A = df[df['Class'] == 'A']
B = df[df['Class'] == 'B']
print(A.shape, B.shape)
```

```python
A_sample = A.sample(n=10)
A_sample.shape
```

```python
new_df = pd.concat([A_sample, B])
new_df
```

```python
new_df['Class'].value_counts()
# now this data is ready for ML
```
