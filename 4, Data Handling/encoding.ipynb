```python
import pandas as pd

df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'City': ['Mumbai', 'Delhi', 'Pune', 'Delhi'],
    'Education': ['Low', 'Medium', 'High', 'Medium']
})



df
```

```python
# 1. Label Encoding (Alphabetical)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Encoded_Gender'] = le.fit_transform(df['Gender'])
df
# this is done by alphabetical order 
# here F comes first & then M 
```

```python
# 2. Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Low', 'Medium', 'High']]) # this is imp step here (order from low to high)
df['Encoded_Education'] = oe.fit_transform(df[['Education']])
df
```

```python
# 3. One Hot encoding
# a) method 1
df_encoded = pd.get_dummies(df, columns=['City'])
df_encoded
```

```python
# b) method 2
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False) 
city_encoded = ohe.fit_transform(df[['City']])
print(city_encoded)

# sparse_output = False 
# all values (0 and 1) are stored (matrix)
# sparse_output = True
# only non-zero (1) values are stored (numpy array)
```

```python
# 4. Dummy encoding
dummy = pd.get_dummies(df, columns=['City'], drop_first=True)
dummy
# all false means last city
```

```python
# 5. Target Encoding
# Replace category with average target value.
data = pd.DataFrame({
    'City': ['Mumbai', 'Delhi', 'Pune', 'Delhi'],
    'Price': [100, 200, 150, 250]
})

city_mean = data.groupby('City')['Price'].mean()
data['City_target_encoded'] = data['City'].map(city_mean)
data
```
