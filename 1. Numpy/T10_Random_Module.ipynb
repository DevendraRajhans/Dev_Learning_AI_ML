```python
import numpy as np
```

```python
# 1. random.rand
print(np.random.rand(5))
print()
print(np.random.rand(4, 2))
```

```python
# 2. random.randn
# It generates random values from (Normal Distribution)
# i.e.  1) mean = 0
#       2) std deviation = 1

print(np.random.randn(5))
print()
print(np.random.randn(4, 2))
```

```python
# 3. random integer
# np.random.randint(low, high, size)

print(np.random.randint(5, 9, 3))
print()
print(np.random.randint(1, 5, (3, 2)))
```

```python
# 4. random.choice

print(np.random.choice([55, 98, 34, 21, 60])) # single choice
print() 
print(np.random.choice([55, 98, 34, 21, 60], size = 4)) # with replacement  
print()
print(np.random.choice([55, 98, 34, 21, 60], size = 4, replace = False)) # without replacement  
```

```python
# 5. random.seed(any_number)
# always same random output for same number 
# numbers -> gives starting point

np.random.seed(46) 
print(np.random.rand(5))

np.random.seed(91)
print(np.random.rand(5))

np.random.seed(46)
print(np.random.rand(5))

np.random.seed(46)
print(np.random.rand(5))

np.random.seed(91)
print(np.random.rand(5))
```
