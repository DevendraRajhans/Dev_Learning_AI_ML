```python
import numpy as np
```

```python
# 1. zeros
# 1d zeros
ar1 = np.zeros(4)
print(ar1)
```

```python
# 2d zeros
ar2 = np.zeros((3, 2))
print(ar2)
```

```python
# 2. ones
# 2D ones
ar3 = np.ones((2, 4))
print(ar3)
```

```python
# 3. full
# 2D full with any value
ar4 = np.full((3, 3), 8) # np.full(shape, value)
print(ar4)
```

```python
# 4. arange
# np.arange(start, stop, step)
# 1D array
ar6 = np.arange(3, 13, 2) # stop in not included
print(ar6)
```

```python
# For 2D or 3D array we use .reshape(shape)
print(np.arange(12))
ar7 = np.arange(12).reshape(3, 4) # imp (total elements = rows*cols)
print(ar7)
print(ar7.reshape(3, 2, 2))
```

```python
# 5. linspace
# creates evenly spaced numbers between a range
# np.linspace(start, stop, no_of_values)
ar8 = np.linspace(0, 10, 4) # stop is included
print(ar8)
```

```python
# 6. identity matrix (always 2D)
# np.eye() 
ar9 = np.eye(4)
print(ar9)
```

```python
# 7. random array creation
# np.random.rand(shape)
# values are in between (0,1)
a = np.random.rand(4)
b = np.random.rand(2, 3)
print("a = ",a)
print("\nb = ",b)
```

```python
# np.random.randn(shape)
# give random values from "Normal distribution"
a = np.random.randn(2, 3) #2D 
print(a)
```

```python
# np.random.randint(low, high, shape)
a = np.random.randint(3, 10, (2, 3)) #2D random integers (in a range 3 to 10)
print(a)
```
