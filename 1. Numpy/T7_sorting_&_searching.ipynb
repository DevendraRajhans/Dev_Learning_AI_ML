```python
import numpy as np
```

```python
a1 = np.array([6, 4, 9, 3, 1])

a2 = np.array([[3, 7, 1],
               [9, 4, 8]])
```

```python
print(np.sort(a1))
print()
print(np.sort(a2)) #by default it sort row-wise (axis = -1)
print()
print(np.sort(a2, axis=0)) # column-wise
print()
print(np.argsort(a1)) # sorting by displaying indexes (use it when ranking is needed)
```

```python
# where
# np.where(condition, value_if_true, value_if_false)
print(np.where(a1 > 5))
print()
print(np.where(a1 > 5, 999, a1)) # if true, 999 && if false, as it is (a1)
```

```python
# index of min & max
print("index of min :", np.argmin(a1))
print("index of max :", np.argmax(a1))
print()
print("index of max of 2D:", np.argmax(a2)) #for 2D array it acts as flatten array by default
# we can apply (axis = 0) & (axis = 1) here also for column-wise & row-wise
```

```python
x = np.array([40, 20, 40, 30, 20, 40, 60, 70])
print(np.unique(x))

# this is very useful
values, counts = np.unique(x, return_counts = True)
print("Values :", values)
print("Counts :", counts)
print("i.e.  (20 = 2 times), (40 = 3 times) & (others = 1 time)")
```
