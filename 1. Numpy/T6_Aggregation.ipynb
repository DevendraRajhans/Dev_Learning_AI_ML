```python
import numpy as np
```

```python
a = np.array([[10, 20, 30],
              [40, 50, 60]])
print(a)
```

```python
print("1. sum :", a.sum())
print("2. mean :", a.mean())
print("3. median :", np.median(a)) # <--- different np.median(a)
print("4. min :", a.min())
print("5. max :", a.max())
print("6. standard deviation :", a.std())
print("7. variance :", a.var())
print("8. index of min :", a.argmin())
print("9. index of max :", a.argmax())
```

```python
# Axis concept
# (axis = 0) --> vertical (column-wise) 
# (axis = 1) --> horizontal (row-wise) 

print("column-wise :", a.sum(axis=0)) 
print("row-wise :", a.sum(axis=1))   # similar for all operations
```
