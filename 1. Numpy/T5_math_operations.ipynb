```python
import numpy as np
```

```python
arr = np.array([10, 40, 20, 50])
print(arr + 10)
print(arr - 5)
print(arr * 3)
print(arr ** 2)
print(arr / 10)
print(arr // 6)
print(arr % 4)
```

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a @ b) # dot product : sum of products (10+40+90+160)
```

```python
print(np.add(a, b))
print(np.subtract(a, b))
print(np.multiply(a, b))
print(np.divide(a, b))
print(np.dot(a,b)) # dot product sum of products (10+40+90+160)
```

```python
c = np.array([4, 16, 9, 64])
print(np.sqrt(c)) # square root
print(np.power(c, 2)) # square
print(np.power(c, 3)) # cube
```

```python
print(np.exp(a)) # (e^x) exponential
print(np.log(a)) # log(x) logarithm
```

```python
d = np.array([1.2, 2.4, 3.7, 4.8])
print(np.round(d)) # rounding
print(np.floor(d)) # down
print(np.ceil(d))  # up
```

```python
# Broadcasting
# numpy expands smaller arrays to match bigger array , so operation can happen

a1 = np.array([10, 20])
a2 = np.array([[10, 20], [30, 40]])

print(a1 + 3) # 3 expands to [3, 3]
print()
print(a2 + a1) # here a1 (1D) array expanded to [[10, 20], [10, 20]]
print()

b = np.array([[10],
              [20]])
print(a2 + b) # here b becomes b = [[10, 10], 
                                  # [20, 20]]
 
```
