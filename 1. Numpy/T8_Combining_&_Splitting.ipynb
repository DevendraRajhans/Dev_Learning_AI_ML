```python
import numpy as np
```

```python
# 1. concatenation
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.concatenate((a, b))) #double parethesis is imp (())
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Column-wise: ")
print(np.concatenate((A, B), axis=0))
print("Row-wise: ")
print(np.concatenate((A, B), axis=1))
```

```python
# 2. stacks
# np.vstack() -- vertical stack
# np.hstack() -- horizontal stack
# np.stack()  -- joins array but create new dimention

print("vstack")
print(np.vstack((A, B)))
print("\nhstack")
print(np.hstack((A, B)))
print("\nstack column wise (adds 3rd dimention):")
print(np.stack((A, B), axis=0))
print("\nstack row wise (adds 3rd dimention):")
print(np.stack((A, B), axis=1))
```

```python
# 3. splitting 
# np.split(array, no_of_parts)
# np.vsplit(array, no_of_parts)
# np.hsplit(array, no_of_parts)

X = np.array([1, 2, 3, 4, 5, 6])
print(np.split(X, 3))
```

```python
Y = np.array([[1, 2, 3, 4], 
              [5, 6, 7, 8]])

print(np.hsplit(Y, 4))
print()
print(np.vsplit(Y, 2))
```
