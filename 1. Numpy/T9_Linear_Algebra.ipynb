```python
import numpy as np
```

```python
# 1. ( dot product = matrix multiplication )
# matrix multiplication
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(A@B)
print()
print(np.dot(A, B))
```

```python
# for 1D 
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b)) # (1*4 + 2*5 + 3*6) = 32
print(a@b)
```

```python
# 2. Determinant
print(np.linalg.det(A))
```

```python
# 3. Inverse
A_inv = np.linalg.inv(A)
print(A_inv)
print("---------------------------------")
print("\n Matrix multiplication of A and A_inv (Identity matrix) :  [it is aprox]\n")
print(A @ A_inv)
```

```python
# 4. Eigen values & Eigen vectors
values, vectors = np.linalg.eig(A)
print("Eigen Values : ")
print(values)
print("\nEigen Vectors : ")
print(vectors)
```

```python
# 5. solving equations :

#   x + y = 5 
#   x - y = 1

A = np.array([[1, 1],
              [1, -1]])

b = np.array([5, 1])

#  solving for (Ax = b)
x = np.linalg.solve(A, b)
print(x)
```
